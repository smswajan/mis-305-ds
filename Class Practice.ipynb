```python
import pandas as pd
fileName = "new_data.csv"
df_autos = pd.read_csv(fileName)
```

```python
df_autos.head(10)
```

```python
header = ['extra','symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height', 'curb', 'engine-type', 'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
```

```python
df_autos.columns = header
```

```python
df_autos.head()
```

```python
df_autos.drop(['extra'], axis=1, inplace=True)
```

```python
df_autos.dtypes
```

```python
df_autos.describe(include='all')
```

```python
df_autos
```

```python
df_autos.dropna(subset = ['price'], axis = 0, inplace = True)
```

```python
df_autos
```

```python
mean = df_autos['normalized-losses'].mean()
mean
```

```python
import numpy as np
```

```python
normalized_losses_2 = df_autos['normalized-losses'].replace(np.nan, mean)
```

```python
normalized_losses_2
```

```python
df_autos['normalized-losses'] = normalized_losses_2
```

```python
df_autos.head(20)
```

```python
df_autos['city-mpg'] = 235/df_autos['city-mpg']
```

```python
df_autos.head(20)
```

```python
df_autos.rename(columns = {'city-mpg':'city-L/100km'}, inplace= True)
```

```python
df_autos.head(5)
```

```python
df_autos['length'].max()
```

```python
df_autos['length']
```

```python
df_autos['width'] = (df_autos['width'] - df_autos['width'].min())/(df_autos['width'].max() - df_autos['width'].min())
```

```python
df_autos['height'] =(df_autos['height']-df_autos['height'].mean())/df_autos['height'].std()
```

```python
import matplotlib.pyplot as plt
y = df_autos['engine-size']
x = df_autos['price']

plt.title('Scatterplot of Engine Size vs Price')
plt.ylabel("Engine Size")
plt.xlabel('Price')

plt.scatter(x, y)
```

```python
df_test= df_autos[['drive-wheels', 'body-style', 'price']]
df_grp = df_test.groupby(['drive-wheels', 'body-style'], as_index = False).mean()
```

```python
df_grp
```

```python
# We know pivot table well on excel. Right? Let's do it on pandas with the pivot function
df_pivot = df_grp.pivot(index = 'drive-wheels', columns='body-style')
df_pivot
```

```python
# Let's create a HeatMap to simplify the pivot table above
plt.pcolor(df_pivot, cmap = 'RdBu')
plt.colorbar()
plt.show()
```
